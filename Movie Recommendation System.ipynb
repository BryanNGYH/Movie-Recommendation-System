```python
#abstract synthx tree
# import ast

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L

def convert_1(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['cast'])
    return L

def convert_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies.genres = movies.genres.apply(convert)
# print(movies.genres[:5])
movies.keywords = movies.keywords.apply(convert)
# print(movies.keywords[:5])
movies.production_companies = movies.production_companies.apply(convert)
# print(movies.product_companies[:5])
movies.spoken_languages = movies.spoken_languages.apply(convert)
# print(movies.spoken_languages[:5])
movies.cast = movies.cast.apply(convert)
# movies.cast[:5]
movies['director'] = movies.crew.apply(convert_director)
```

```python
from datetime import datetime
from dateutil.parser import *
```

```python
def extract_date(date_string):
    
    date = datetime.strptime(date_string,'%Y-%m-%d')
    
    year = date.year
    month = date.month
    day = date.day
    
    return year,month,day
```

```python
ab = extract_date('2011-12-11')
print(ab)
a = pd.Series(extract_date('2011-12-11'))
print(a)
```

```python
def remove_space(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```
